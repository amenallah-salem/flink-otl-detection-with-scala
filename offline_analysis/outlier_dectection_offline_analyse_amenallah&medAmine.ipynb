# Let's detect our patern
## working with clickStream_df



First method direct using the fact thet the  uid must be ==ip adress
    
**clickStream_df preprocessing**

```python
def clickStream_df_preprocessing(df_):
    #we first drop the duplicated rows based on their uid
    df_=df_.drop_duplicates(subset = ["uid"], keep ="first")
    #put all unique ip adresses in a np array (using the unique method)!
    uniqueIp = df_['ip'].unique()
    #display uniques ip adresses 
    print('the unisue IP adresses in our dataset are ', uniqueIp)
    print('********************************************************')
    print('the data set cleaned')
    return df_
```

```python
clickStream_df_preprocessing(clickStream_df)
```

```python
def outlier_detection(df_):
    if len(df_)==len(df_['ip'].unique()):
        print("the unique Ip adresses is egual to unique uid's")
    else: 
        print("there are {} unique Ip within {} uid's rows in the {}, some IP's are repeated.".format(len(df_['ip'].unique()),len(df_), df_.name ) )



outlier_detection(clickStream_df)
```
